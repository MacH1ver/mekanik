```python
import numpy as np
```

```python
alpha_steel = 12e-6  # 1/K
alpha_aluminum = 23.5e-6  # 1/K

T_max = 85 + 273.15  # K
T_min = 21 + 273.15  # K
delta_T = T_max - T_min  # K

nominal_dia = 95e-3  # m

tol_upper = 0 # m
tol_lower = -8e-6  # m
```

```python
thermal_expansion = nominal_dia * (alpha_aluminum - alpha_steel) * delta_T

print(f"Thermal expansion difference: {thermal_expansion*1e3} mm")
```

```python
lower_bound_bearing = (nominal_dia + tol_lower)  # m  
upper_bound_bearing = (nominal_dia + tol_upper)  # m
```

```python
print(f'Smallest diameter for interference: {(lower_bound_bearing - thermal_expansion)*1e3} mm')
```

```python
deviation = nominal_dia - (lower_bound_bearing - thermal_expansion)
print(f'Max interference fit at max temp: {deviation*1e6} um')
```

```python
interference_at_max_temp = 0.007e-3  # m

print(f'upper tolerance: {(lower_bound_bearing - thermal_expansion - interference_at_max_temp)*1e3}mm')
```

```python
upper_tolerance = lower_bound_bearing - thermal_expansion - interference_at_max_temp
upper_tolerance*1e3
```
